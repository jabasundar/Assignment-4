```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("=== TELANGANA PDS EDA STARTED ===")

# =========================
# 1. LOAD DATA
# =========================
df = pd.read_csv("../data/processed/master_pds_data.csv")

print("Shape:", df.shape)
display(df.head())

# =========================
# 2. BASIC INFO
# =========================
print("\n=== DATA INFO ===")
df.info()

print("\n=== MISSING VALUES ===")
print(df.isnull().sum().sort_values(ascending=False).head(15))

print("\n=== DUPLICATES ===")
print(df.duplicated().sum())

# =========================
# 3. FEATURE ENGINEERING
# =========================
print("\n=== FEATURE ENGINEERING ===")

# Avoid division by zero
df["utilization_ratio"] = df["nooftrans"] / (df["noofrcs"] + 1)

if "othershoptrans" in df.columns:
    df["portability_ratio"] = df["othershoptrans"] / (df["nooftrans"] + 1)

if "riceafsc" in df.columns:
    df["rice_intensity"] = df["riceafsc"] / (df["nooftrans"] + 1)

print("Features created successfully")

# =========================
# 4. DISTRICT ANALYSIS
# =========================
print("\n=== DISTRICT ANALYSIS ===")

district_top = df.groupby("distname_x")["nooftrans"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))
district_top.plot(kind="bar")
plt.title("Top 10 Districts by Average Transactions")
plt.ylabel("Avg Transactions")
plt.xticks(rotation=45)
plt.show()

# =========================
# 5. MONTHLY TREND
# =========================
print("\n=== MONTHLY TREND ===")

monthly_trend = df.groupby("month_x")["nooftrans"].sum()

plt.figure(figsize=(10,5))
monthly_trend.plot(marker="o")
plt.title("Monthly Transaction Trend")
plt.xlabel("Month")
plt.ylabel("Total Transactions")
plt.grid()
plt.show()

# =========================
# 6. UTILIZATION DISTRIBUTION
# =========================
print("\n=== UTILIZATION DISTRIBUTION ===")

plt.figure(figsize=(8,5))
df["utilization_ratio"].hist(bins=50)
plt.title("Utilization Ratio Distribution")
plt.xlabel("Utilization Ratio")
plt.ylabel("Frequency")
plt.show()

# =========================
# 7. CORRELATION HEATMAP
# =========================
print("\n=== CORRELATION HEATMAP ===")

plt.figure(figsize=(12,8))
corr = df.corr(numeric_only=True)

sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()

# =========================
# 8. OUTLIER DETECTION
# =========================
print("\n=== OUTLIER ANALYSIS ===")

threshold = df["utilization_ratio"].quantile(0.99)
outliers = df[df["utilization_ratio"] > threshold]

print("Outlier Shops Count:", outliers["shopno"].nunique())
display(outliers.head())

# =========================
# 9. SHOP DISTRIBUTION
# =========================
print("\n=== SHOP ACTIVITY DISTRIBUTION ===")

plt.figure(figsize=(8,5))
df["nooftrans"].hist(bins=50)
plt.title("Transaction Distribution Across Shops")
plt.xlabel("No of Transactions")
plt.ylabel("Frequency")
plt.show()

# =========================
# END
# =========================
print("\n=== EDA COMPLETED SUCCESSFULLY ===")
```
