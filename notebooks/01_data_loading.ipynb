```python
import pandas as pd
import glob
import os

print("=== Telangana PDS Analytics ===")

transaction_path = r"../data/raw/transactions"

transaction_files = glob.glob(transaction_path + r"/**/*.csv", recursive=True)

print("Transaction Files Found:", len(transaction_files))

transaction_dfs = []

for file in transaction_files:
    
    print("Loading:", file)
    
    df = pd.read_csv(file)
    
    # Extract year from folder
    year = file.split("/")[-2]
    
    df["year"] = year
    
    transaction_dfs.append(df)

transactions = pd.concat(transaction_dfs, ignore_index=True)

print("Transactions Shape:", transactions.shape)

transactions.head()

card_path = r"../data/raw/card_status"

card_files = glob.glob(card_path + r"/**/*.csv", recursive=True)

print("Card Files Found:", len(card_files))

card_dfs = []

for file in card_files:
    
    print("Loading:", file)
    
    df = pd.read_csv(file)
    
    year = file.split("/")[-2]
    
    df["year"] = year
    
    card_dfs.append(df)

card_status = pd.concat(card_dfs, ignore_index=True)

print("Card Status Shape:", card_status.shape)

card_status.head()

import glob

fps_files = glob.glob("../data/raw/fps_locations/*.csv")

print("FPS Files Found:", fps_files)

fps_locations = pd.read_csv(fps_files[0])

print("FPS Shape:", fps_locations.shape)

fps_locations.head()

print(transactions.columns)

print(card_status.columns)

print(fps_locations.columns)

```
